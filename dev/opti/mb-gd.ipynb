```python
import numpy as np
```

```python
x = np.arange(-5, 5, 0.2)
```

```python
y = lambda x: (x)**2
```

```python
dy = lambda x: 2*(x)
```

```python
plot(x, y(x))
```

```python
start = -4.5
```

```python
plot(x, y(x))
plot(start, y(start), 'o')
```

```python
grad = dy(start)
```

```python
grad
```

```python
rate = 0.3
```

```python
second = start - rate * dy(start)
```

```python
plot(x, y(x))
plot(start, y(start), 'o')
plot(second, y(second), 'o')
```

```python
prev = start
rate = 0.1
```

```python
new = prev - rate * dy(prev)
plot(x, y(x))
plot(start, y(start), 'o')
plot(second, y(second), 'o')
plot(new, y(new), 'o')
prev = new
```
