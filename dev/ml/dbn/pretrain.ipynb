```python
import numpy as np
import pandas as pd
from copper.ml.dbn import DBN
from sklearn.metrics import accuracy_score
```

```python
import cPickle, gzip, numpy
f = gzip.open('mnist.pkl.gz', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()
X_train, y_train = train_set[0], train_set[1]
X_valid, y_valid = valid_set[0], valid_set[1]
```

```python
acc = []
def cb(dbn, layer, epoch, j):
    acc.append((dbn.predict(X_train) == y_train).mean())
```

```python
acc
```

```python
dbn = DBN([25, 25, 25], pretrain_epochs=100, pretrain_batch_size=2000, finetune_epochs=0, pretrain_callback=cb, progress_bars=True)
```

```python
dbn.fit(X_train, y_train)
```

```python
accuracy_score(y_valid, dbn.predict(X_valid))
```
