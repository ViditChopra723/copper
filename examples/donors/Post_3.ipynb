We import the data and set some roles for the variables

```python
import copper
copper.config.path = '../../tests'
ds = copper.read_csv('donors/data.csv')
ds.role['TARGET_D'] = ds.REJECTED
ds.role['TARGET_B'] = ds.TARGET
ds.type['ID'] = ds.CATEGORY
```

Since scikit-learn cant handle NANs we need to fill the values so I create simple method con the Dataset class to fill values of numerical columns with the mean.

```python
ds.fillna('DemAge', 'mean')
ds.fillna('GiftAvgCard36', 'mean')
```

Let's see if we are good

```python
ds.inputs
```

OK, no missing values so we are good to go.

## Machine Learning

Time to see if boosting and bagging are that good as they promise to be. We create a new machine learning instance and set the dataset and we tell them to sample half for training and half for testing.

```python
ml = copper.MachineLearning()
ml.dataset = ds
ml.sample(0.5)
```

Create a new Decision Tree, add it to the models to compare and fit the models

```python
from sklearn import tree
tree_clf = tree.DecisionTreeClassifier(max_depth=10)
ml.add_model(tree_clf, 'Decision Tree')
ml.fit()
```

Since I start coding this library I wanted to maintain all compatibility with pandas and scikit-learn and its respective API. For that reason is possible to just add already fitted classifiers to the ML class and then just compare them with the utilities; this is useful in the case of using bootstraping.

In the next lines I create 5 different Decision Tree classifiers using 5 different samples (using Bootstrap), fit each classifier and add them to the ML class.

```python
from sklearn import cross_validation
bs = cross_validation.Bootstrap(len(ds.inputs.values), n_iter=5)
i = 0
for train_index, test_index in bs:
    X_train = ds.inputs.values[test_index]
    y_train = ds.target.values[test_index]
    clf = tree.DecisionTreeClassifier(max_depth=10)
    clf.fit(X_train, y_train)
    ml.add_model(clf, "DT" + str(i + 1))
    i += 1
```

Lets see some results

```python
ml.accuracy()
```

```python
ml.roc()
```

I was amazed to see that just using bootstraping to sample instead of a simple sample generates better results. The first model is barely above the random line so it sucks, but the other models are better.

### Bagging

Since scikit-learn does not have a implementation of bagging. I made a really simple (and un-efficient) implementation of bagging by using the mode to predict the class and using the mean to predict the probabilities. Finally create some methods to make that possible and easy on the ML class.

But first I want to remove the first Decision Tree model, because it sucks.

```python
ml.remove_model('Decision Tree')
```

```python
ml.models # Checking is gone
```

To create a bag of all models is as simple as call the method and pass a name as parameter. On future releases will be possible to pass a list of target models to use on the bag and that way will be easy to create more bags.

```python
ml.bagging("Bagging")
```

Let's see the results

```python
ml.accuracy()
```

```python
ml.roc()
```

That is a big improvement on both accuracy and ROC chart.
