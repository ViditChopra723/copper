```python
import copper
import pandas as pd
```

```python
train = copper.read_csv('train.csv')
train.role['depend'] = train.TARGET
train.fillna(method='mean')
```

```python
test = copper.read_csv('test.csv')
test.role['depend'] = test.TARGET
test.fillna(method='mean')
```

```python
ml = copper.MachineLearning()
ml.train = train
ml.test = test
```

```python
ml.clear_clfs()
ml.clfs
```

```python
from sklearn import tree
tree_clf = tree.DecisionTreeClassifier(max_depth=6)
tree_clf.fit(ml.X_train, ml.y_train)
ml.add_clf(tree_clf, 'Decision Tree')
```

```python
from sklearn import svm
svc = svm.SVC(kernel='linear', C=1)
ml.add_clf(svc, 'SVM')
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(loss="log", penalty="l2")
sgd_clf.fit(ml.X_train, ml.y_train)
ml.add_clf(sgd_clf, "SGD")
```

```python
from sklearn.naive_bayes import GaussianNB
gnb_clf = GaussianNB()
gnb_clf.fit(ml.X_train, ml.y_train)
ml.add_clf(gnb_clf, "GNB")
```

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid
nc_clf = NearestCentroid()
nc_clf.fit(ml.X_train, ml.y_train)
ml.add_clf(nc_clf, "NC")
```

```python
ml.rm_clf('NC')
```
