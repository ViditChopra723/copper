```python
import re
import numpy as np
import pandas as pd
```

```python
loans = pd.read_csv('../data/loansData.csv')
```

```python
loans.head()
```

```python
p = re.compile('[0-9.]+')
def number(x):
    try:
        return float(p.search(x).group())
    except: 
        return 0
```

```python
percent = lambda x: number(x) / 100
loans['Interest.Rate'] = loans['Interest.Rate'].apply(percent)
```

```python
loans['Loan.Length'] = loans['Loan.Length'].apply(number)
```

```python
loans['Debt.To.Income.Ratio'] = loans['Debt.To.Income.Ratio'].apply(number)
```

```python
loans['Employment.Length'] = loans['Employment.Length'].apply(number)
```

```python
loans.head(10)
```

```python
fico_init = lambda x: int(x.split('-')[0])
fico_final = lambda x: int(x.split('-')[1])
loans['FICO.Init'] = loans['FICO.Range'].apply(fico_init)
loans['FICO.Final'] = loans['FICO.Range'].apply(fico_final)
```
