```python
import numpy as np
import sklearn.datasets as datasets
from lr import LogisticRegression
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
y[y == 1] = 5
```

```python
class LogisticRegressionClassifier(object):
    
    def __init__(self, *args, **kwargs):
        self.clfs = {}
        self.labels = ()
        self.args = args
        self.kwargs = kwargs
    
    def fit(self, X, y):
        self.labels = set(y)
        for label in self.labels:
            lr = LogisticRegression(*self.args, **self.kwargs)
            lr.fit(X, y == label)
            self.clfs[label] = lr
        
    def predict_proba(self, X):
        ans = np.zeros((X.shape[0], len(self.labels)))
        for i, label in enumerate(self.labels):
            clf = self.clfs[label]
            ans[:, i] = clf.predict_proba(X)
        return ans
    
    def predict(self, X):
        max_indexes = self.predict_proba(X).argmax(axis=1)
        ans = -1 * np.ones(X.shape[0])
        for i, label in enumerate(self.labels):
            ans[max_indexes == i] = label
        return ans
```

```python
lrc = LogisticRegressionClassifier(reg_lambda=0)
lrc.fit(X, y)
```

```python
lrc.predict_proba(X)
```

```python
lrc.predict(X)
```

```python
(lrc.predict(X) == y).mean()
```

```python
%timeit lrc.fit(X, y)
```
